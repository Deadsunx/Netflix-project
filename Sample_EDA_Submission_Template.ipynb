## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.

# ---- 1. Handle missing values ----
df['director'] = df['director'].fillna('Unknown')
df['cast'] = df['cast'].fillna('Unknown')
df['country'] = df['country'].fillna('Unknown')
df['rating'] = df['rating'].fillna(df['rating'].mode()[0])

# Drop rows where 'date_added' is missing (only 10 rows)
df = df.dropna(subset=['date_added'])
print(f'Shape after handling missing values: {df.shape}')
print(f'Remaining nulls: {df.isnull().sum().sum()}')

# ---- 2. Parse date_added into datetime ----
df['date_added'] = pd.to_datetime(df['date_added'].str.strip())
df['year_added'] = df['date_added'].dt.year
df['month_added'] = df['date_added'].dt.month
df['month_name_added'] = df['date_added'].dt.month_name()
df['day_of_week_added'] = df['date_added'].dt.day_name()

# ---- 3. Extract numeric duration ----
df['duration_int'] = df['duration'].str.extract(r'(\d+)').astype(int)

# ---- 4. Extract the primary country ----
df['primary_country'] = df['country'].apply(lambda x: str(x).split(',')[0].strip())

# ---- 5. Extract individual genres ----
df['genres'] = df['listed_in'].apply(lambda x: [g.strip() for g in str(x).split(',')])

# Create separate dataframes for Movies and TV Shows (after all feature engineering)
movies = df[df['type'] == 'Movie'].copy()
tv_shows = df[df['type'] == 'TV Show'].copy()

print(f'\nFinal dataset shape: {df.shape}')
print(f'New columns: year_added, month_added, month_name_added, day_of_week_added, duration_int, primary_country, genres')
print(f'\nMissing values after wrangling:')
print(df.isnull().sum())
````

### What all manipulations have you done and insights you found?

**Data Wrangling Summary:**

1. **Handled Missing Values:**
   - Filled `director`, `cast`, and `country` nulls with 'Unknown' to avoid data loss.
   - Filled `rating` nulls with the mode value ('TV-MA').
   - Dropped 10 rows where `date_added` was null (negligible data loss).

2. **Parsed `date_added`** into proper datetime, then extracted: `year_added`, `month_added`, `month_name_added`, `day_of_week_added`.

3. **Extracted numeric duration** (`duration_int`) -- minutes for movies, seasons for TV Shows.

4. **Extracted primary country** from multi-country entries.

5. **Split genres** from `listed_in` into a list per title.

6. **Created separate DataFrames** for Movies and TV Shows.
